```python
import polars as pl
import hvplot.polars
```

```python
df = pl.read_csv("financial_sentiment_analysis.csv")
df = df.unique()
```

```python
df.filter(pl.col('Positive') > 0.9).group_by('date').len().sort('date').hvplot.line(x='date', y='len', title='Number of Positive Articles Over Time', height=400, width=800)
```

```python
df.filter(pl.col('Negative') > 0.97).group_by('date').len().sort('date').hvplot.line(x='date', y='len', title='Number of Negative Articles Over Time', height=400, width=800)
```

```python
df.filter(pl.col('Neutral') > 0.75).group_by('date').len().sort('date').hvplot.line(x='date', y='len', title='Number of Neutral Articles Over Time', height=400, width=800)
```

```python
df.group_by('date').len().hvplot(x = 'date', y = 'len')
```

```python
df.group_by('date').len('comnam').hvplot(x = 'date', y = 'comnam')
```
